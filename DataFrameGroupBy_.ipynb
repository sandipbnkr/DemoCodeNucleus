```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f=pd.read_csv('https://raw.githubusercontent.com/hakimyameen/Datasets/main/fifa_eda.csv')
```

```python
f.head()
```

```python
#group by using pands
#applied on categorical data
#select categorical data
f.select_dtypes('object')
```

```python
#performing groupby
f.groupby('Nationality')
```

```python
#it gives u object then typecast it to groupings
list(f.groupby('Nationality'))
```

```python
#when we perform a groupby it gives 2 things: category and categorical data
```

```python
list(f.groupby('Nationality'))
```

```python
#if we want to get a data of perticular category
f.groupby('Nationality').get_group('India')
#this will return data(dataframe)
```

```python
#fetc fist record of each country
f.groupby('Nationality').first()
```

```python
#fetch last record of each categoory
f.groupby('Nationality').last()
```

```python
#fetch only perticular columns into dataframe
f.groupby('Nationality').last()[['Name','Club']]
```

```python
f[['Nationality','Preferred Foot']].describe()
```

```python
#lets use 2 coloumbs with groupby: it gives first left and then right as ascending orde form
list(f.groupby(['Nationality','Preferred Foot']))
```

```python
f.groupby(['Nationality','Preferred Foot']).Age.mean()
```

```python
#groupby gives u aggregate facility
#max age of each country with aggregate operation
f.groupby('Nationality').Age.agg(max)
```

```python
f.groupby('Nationality')[['Age','Wage']].agg(max)
```

```python
#can we apply more aggr functions
f.groupby('Nationality')[['Age','Wage']].agg([max,min,np.mean])
```

```python
#sorting
f.groupby('Nationality'.first().sort_values(by='Joined')
```

```python
#usie this approch to sort than above one
fs=f.groupby('Nationality').first()
fs.sort_values(by='Joined')
```
